```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
# Seed the random function to generate same data every time.
np.random.seed(0)
# Randomly generate a set of evenly distributed data consist of nrow sets of numbers ( each set contains ncol values )
# with value between -1 and 1 ( 2-D array )
matrix = np.random.uniform(-1,1,[nrow,ncol])
print (matrix)

# 隨機給予 0, 1, 2 三種標籤
# Randomly assign a value selected from a list
indice = np.random.choice([0,1,2], size=nrow)
# Create a data frame using the matrix and assign row number using the indice
plot_data = pd.DataFrame(matrix, indice)
# plot_data

# 繪製 seaborn 進階 Heatmap
# Create pair grid object
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
# Use scatter for upper part of the pair grid
grid.map_upper(plt.scatter, alpha = 0.2)

# Use histogram for diagonal line of the pair grid
grid.map_diag(sns.kdeplot)

# Use density plot for lower part of the pair grid
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

# Set overall title
plt.suptitle('Index and Random Data Pairs Plot', size = 32, y = 1.05)

# Show chart
plt.show()

# The distribution should spread evenly
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
# Seed the random function to generate same data every time.
np.random.seed(0)
# Randomly generate a set of normally data consist of nrow sets of numbers ( each set contains ncol values )
# with medium value 1 and standard deviation 0.2
matrix = np.random.normal(1, 0.2,[nrow,ncol])


# Randomly assign a value selected from a list
indice = np.random.choice([0,1,2], size=nrow)
# Create a data frame using the matrix and assign row number using the indice
plot_data = pd.DataFrame(matrix, indice)

"""
Your Code Here
Please replace "..." to correct plot function
"""
# Create pair grid object
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
# Use scatter for upper part of the pair grid
grid.map_upper(plt.scatter, alpha = 0.2)

# Use histogram for diagonal line of the pair grid
grid.map_diag(sns.kdeplot)

# Use density plot for lower part of the pair grid
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

# Set overall title
plt.suptitle('Index and Random Data Pairs Plot', size = 32, y = 1.05)

# Show chart
plt.show()

# The distribution should focus mostly in the center
```
