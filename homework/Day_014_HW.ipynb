```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# Set data directory
dir_data = 'D:\Document\AI\Marathon100D\Assignment\Day_014\data'
```

```python
# Set the full data file name
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))

# Read the csv data file
app_train = pd.read_csv(f_app)

# Print first few rows ( default to be 5 rows) of data
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
# Remove rows with null column value and extract just one column
unique_house_type = app_train[app_train['HOUSETYPE_MODE'].notnull()]["HOUSETYPE_MODE"].unique()
print (unique_house_type)
print (app_train['HOUSETYPE_MODE'].value_counts())

# Get subplot number of rows using number of unique value
nrows = len(unique_house_type)

# Set subplot number of columns using half of the rows
ncols = nrows // 2

# Set chart size
plt.figure(figsize=(10,30))

for i in range(len(unique_house_type)):
    # Set position of the subplot
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    # Draw histogram subplot
    app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]), 'AMT_CREDIT'].hist( bins=20, grid=False, color='#86bf91', zorder=2, rwidth=0.9)   
    plt.title(str(unique_house_type[i]))
    
plt.show()
```
