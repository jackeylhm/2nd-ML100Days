# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):
    return abs(x) * (x > 0)

# linespace generate an array from start and stop value with requested number of elements.
# Generate 100 numbers with equal space ranging from -10 to 10
x = plt.linspace(-10,10,100)

# Prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# Set plot title
plt.title('ReLu function')

# Show Relu formula
plt.text(4, 0.8, r'abs(x) * (x > 0)', fontsize=15)

# Resize X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
#plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 
# Plot graph
plt.show()
```

```python
def dReLU(x):
    return 1. * (x > 0)

# linespace generate an array from start and stop value with requested number of elements.
# Generate 100 numbers with equal space ranging from -10 to 10
x = plt.linspace(-10,10,100)

# Prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, dReLU(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# Set plot title
plt.title('dReLu function')

# Show Relu formula
plt.text(4, 0.8, r'1. * (x > 0)', fontsize=15)

# Resize X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
#plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 
# Plot graph
plt.show()
```
