### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
# Assign result of data frame row count 
rowCount = df_app_train.shape[0]  # len(df_app_train)
print('Row count of file is %s' % (rowCount))

# Assign result of data frame column count 
colCount = df_app_train.shape[1]  # len(df_app_train.columns)
print('Column count of file is %s' % (colCount))
```

#### 列出所有欄位

```python
print ('Listing all column names of data frame')
list(df_app_train)    # df_app_train.columns
```

#### 截取部分資料

```python
# Show first 5 rows of data frame
df_app_train.head()

# Show last 5 rows of data frame
df_app_train.tail()

# Show 10 random samples of data frame
df_app_train.sample(n=10, random_state=1)

# Show 0.1 percent random samples of data frame
df_app_train.sample(frac=0.001, random_state=1)
```
